```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# Download dataset
import kagglehub
path = kagglehub.dataset_download("gokulrajkmv/unemployment-in-india")
print("Path to dataset files:", path)
```

 DATA LOADING & EXPLORATION

```python
print("\n" + "="*60)
print("UNEMPLOYMENT DATA ANALYSIS PROJECT")
print("="*60)

# Load the unemployment data
df = pd.read_csv(f"{path}/Unemployment in India.csv")

# Display basic information
print("\n1. DATASET OVERVIEW")
print("-" * 60)
print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
print(f"\nColumn names and types:")
print(df.dtypes)
print(f"\nMissing values:")
print(df.isnull().sum())
print(f"\nBasic statistics:")
print(df.describe())
```

DATA CLEANING

```python
print("\n2. DATA CLEANING")
print("-" * 60)

# Make a copy for cleaning
df_clean = df.copy()

# Standardize column names
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(' ', '_')
print(f"Standardized columns: {df_clean.columns.tolist()}")
```

```python
# Convert date column
date_columns = [col for col in df_clean.columns if 'date' in col]
if date_columns:
    df_clean[date_columns[0]] = pd.to_datetime(df_clean[date_columns[0]])
    df_clean = df_clean.sort_values(by=date_columns[0])
    print(f"Date range: {df_clean[date_columns[0]].min()} to {df_clean[date_columns[0]].max()}")

```

```python
# Handle missing values
if df_clean.isnull().sum().sum() > 0:
    df_clean = df_clean.dropna()
    print(f"Removed rows with missing values. New shape: {df_clean.shape}")
```

```python
# Identify unemployment rate column
unemployment_col = [col for col in df_clean.columns if 'unemployment' in col or 'rate' in col.lower()]
unemployment_col = unemployment_col[0] if unemployment_col else df_clean.columns[-1]
print(f"Unemployment rate column: {unemployment_col}")

print(f"\nCleaned data shape: {df_clean.shape}")
print("Data cleaning completed!")
```

OVID-19 IMPACT ANALYSIS

```python
print("\n3. COVID-19 IMPACT ANALYSIS")
print("-" * 60)

date_col = date_columns[0]
covid_start = pd.to_datetime('2020-03-01')
covid_severe = pd.to_datetime('2020-04-01')
covid_recovery = pd.to_datetime('2020-09-01')
vaccine_start = pd.to_datetime('2021-01-01')
```

```python
# Create periods
pre_covid = df_clean[df_clean[date_col] < covid_start]
covid_severe_period = df_clean[(df_clean[date_col] >= covid_start) & (df_clean[date_col] < covid_recovery)]
recovery_period = df_clean[(df_clean[date_col] >= covid_recovery) & (df_clean[date_col] < vaccine_start)]
post_vaccine = df_clean[df_clean[date_col] >= vaccine_start]

print(f"Pre-COVID (before Mar 2020) - Avg Unemployment: {pre_covid[unemployment_col].mean():.2f}%")
print(f"COVID Severe (Mar-Aug 2020) - Avg Unemployment: {covid_severe_period[unemployment_col].mean():.2f}%")
print(f"Recovery Phase (Sep-Dec 2020) - Avg Unemployment: {recovery_period[unemployment_col].mean():.2f}%")
print(f"Post-Vaccine (Jan 2021+) - Avg Unemployment: {post_vaccine[unemployment_col].mean():.2f}%")
```

```python
# Calculate impact
if len(pre_covid) > 0 and len(covid_severe_period) > 0:
    covid_impact = covid_severe_period[unemployment_col].mean() - pre_covid[unemployment_col].mean()
    print(f"\nCOVID-19 Impact: +{covid_impact:.2f}% increase in unemployment")
```

SEASONAL TRENDS

```python
print("\n4. SEASONAL TRENDS ANALYSIS")
print("-" * 60)

df_clean['month'] = df_clean[date_col].dt.month
df_clean['year'] = df_clean[date_col].dt.year
df_clean['quarter'] = df_clean[date_col].dt.quarter

seasonal_monthly = df_clean.groupby('month')[unemployment_col].agg(['mean', 'std', 'min', 'max'])
seasonal_quarterly = df_clean.groupby('quarter')[unemployment_col].agg(['mean', 'std'])

print("Monthly Average Unemployment Rates:")
print(seasonal_monthly)
print("\nQuarterly Average Unemployment Rates:")
print(seasonal_quarterly)
```

STATE-WISE ANALYSIS (if available)

```python
print("\n5. STATE/REGION-WISE ANALYSIS (if available)")
print("-" * 60)

state_cols = [col for col in df_clean.columns if 'state' in col or 'region' in col]
if state_cols:
    state_col = state_cols[0]
    state_analysis = df_clean.groupby(state_col)[unemployment_col].agg(['mean', 'std', 'min', 'max']).sort_values('mean', ascending=False)
    print(f"\nTop 10 states/regions with highest unemployment:")
    print(state_analysis.head(10))
else:
    print("No state/region column found in dataset")
```

VISUALIZATIONS

```python
print("\n6. GENERATING VISUALIZATIONS")
print("-" * 60)

fig = plt.figure(figsize=(16, 12))

# Plot 1: Unemployment trend over time
ax1 = plt.subplot(3, 2, 1)
ax1.plot(df_clean[date_col], df_clean[unemployment_col], linewidth=2, color='#1f77b4')
ax1.axvspan(covid_start, covid_recovery, alpha=0.2, color='red', label='COVID-19 Severe Period')
ax1.axvline(vaccine_start, color='green', linestyle='--', linewidth=2, label='Vaccine Start')
ax1.set_title('Unemployment Rate Trend Over Time', fontsize=12, fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('Unemployment Rate (%)')
ax1.legend()
ax1.grid(True, alpha=0.3)
```

```python
# Plot 2: COVID-19 Impact Comparison
ax2 = plt.subplot(3, 2, 2)
periods = ['Pre-COVID', 'COVID Severe', 'Recovery', 'Post-Vaccine']
values = [
    pre_covid[unemployment_col].mean() if len(pre_covid) > 0 else 0,
    covid_severe_period[unemployment_col].mean() if len(covid_severe_period) > 0 else 0,
    recovery_period[unemployment_col].mean() if len(recovery_period) > 0 else 0,
    post_vaccine[unemployment_col].mean() if len(post_vaccine) > 0 else 0
]
colors = ['green', 'red', 'orange', 'blue']
bars = ax2.bar(periods, values, color=colors, alpha=0.7)
ax2.set_title('COVID-19 Impact on Unemployment', fontsize=12, fontweight='bold')
ax2.set_ylabel('Average Unemployment Rate (%)')
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%', ha='center', va='bottom', fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
```

```python
# Plot 3: Monthly Seasonality
ax3 = plt.subplot(3, 2, 3)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax3.plot(range(1, 13), seasonal_monthly['mean'], marker='o', linewidth=2, markersize=8, color='#ff7f0e')
ax3.fill_between(range(1, 13), seasonal_monthly['mean'] - seasonal_monthly['std'],
                  seasonal_monthly['mean'] + seasonal_monthly['std'], alpha=0.3, color='#ff7f0e')
ax3.set_title('Seasonal Patterns - Monthly Average', fontsize=12, fontweight='bold')
ax3.set_xlabel('Month')
ax3.set_ylabel('Unemployment Rate (%)')
ax3.set_xticks(range(1, 13))
ax3.set_xticklabels(months)
ax3.grid(True, alpha=0.3)
```

```python
# Plot 4: Quarterly Seasonality
ax4 = plt.subplot(3, 2, 4)
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
ax4.bar(quarters, seasonal_quarterly['mean'], color=['#2ca02c', '#d62728', '#9467bd', '#8c564b'], alpha=0.7)
ax4.set_title('Seasonal Patterns - Quarterly Average', fontsize=12, fontweight='bold')
ax4.set_ylabel('Unemployment Rate (%)')
ax4.grid(True, alpha=0.3, axis='y')
```

```python
# Plot 5: Distribution
ax5 = plt.subplot(3, 2, 5)
ax5.hist(df_clean[unemployment_col], bins=30, color='#1f77b4', alpha=0.7, edgecolor='black')
ax5.axvline(df_clean[unemployment_col].mean(), color='red', linestyle='--', linewidth=2, label=f"Mean: {df_clean[unemployment_col].mean():.2f}%")
ax5.axvline(df_clean[unemployment_col].median(), color='green', linestyle='--', linewidth=2, label=f"Median: {df_clean[unemployment_col].median():.2f}%")
ax5.set_title('Unemployment Rate Distribution', fontsize=12, fontweight='bold')
ax5.set_xlabel('Unemployment Rate (%)')
ax5.set_ylabel('Frequency')
ax5.legend()
ax5.grid(True, alpha=0.3, axis='y')
```

```python
# Plot 6: Year-over-year comparison
ax6 = plt.subplot(3, 2, 6)
yearly_avg = df_clean.groupby('year')[unemployment_col].mean()
ax6.bar(yearly_avg.index, yearly_avg.values, color='#17becf', alpha=0.7, edgecolor='black')
ax6.set_title('Year-over-Year Unemployment Average', fontsize=12, fontweight='bold')
ax6.set_xlabel('Year')
ax6.set_ylabel('Average Unemployment Rate (%)')
ax6.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('unemployment_analysis.png', dpi=300, bbox_inches='tight')
print("✓ Visualization saved as 'unemployment_analysis.png'")
plt.show()
```
